```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
data = [
    {"Period": "Early Development", "Start": -7240, "End": -3500, "Event": "Early farming in the Andes (-7,240 to -3,500 BCE)"},
    {"Period": "Prehispanic Agriculture", "Start": -2000, "End": 1400, "Event": "Intensive irrigation-based farming (-2,000 BCE to 1,400 CE)"},
    {"Period": "Colonial Period", "Start": 1681, "End": 1750, "Event": "Economic crisis and agricultural decline (1,675 to 1750 CE)"},
    {"Period": "Colonial Period", "Start": 1750, "End": 1800, "Event": "Agricultural growth and expansion (1,750 to 1,800 CE)"},
    {"Period": "Republican Era", "Start": 1875, "End": 1933, "Event": "Rise of plantation agriculture (1,875 to 1,933 CE)"},
    {"Period": "Agriculture in Peru Today", "Start": 1875, "End": 1933, "Event": "Agricultural Economic Growth (1,994 to 2022 CE)"},
]

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the linear timeline plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the events
for i, row in df.iterrows():
    ax.plot([row["Start"], row["End"]], [i, i], marker="o", markersize=8, label=row["Event"])
    ax.text(row["Start"], i, row["Period"], va="center", ha="right", fontsize=10)

# Customize plot
ax.set_xlim(-10000, 2000)
ax.set_ylim(-1, len(df))
ax.set_yticks(range(len(df)))
ax.set_yticklabels([])  # Hide y-axis labels
ax.set_xlabel("Year")
ax.set_title("Agricultural Timeline")
ax.grid(axis="x", linestyle="--", alpha=0.7)

# Add legend
ax.legend(loc="upper left")

plt.tight_layout()
plt.show()
```
