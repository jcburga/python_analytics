# Agriculture in Peru:
    # Main Indicators
        # Total Revenue   
        # Agriculture contribution to the National GDP
        # Agriculture Added Value
        # Crops, Land Dedicated to...
        # Harvested Land

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r'C:\Users\jcbur\OneDrive - peruvianbusinesscouncil.com\csv files\#11_python_final_assignment\#3_bottomline_csv_ datasets\agriculture_indicators.csv')

# Define categories for visualization
categories = ['revenue_pen_2007', 'gdp_percent', 'added_value_PEN07', 'harvested_mill_ha', 'crops_mill_ha']

# Plotting each category
for category in categories:
    plt.figure(figsize=(10, 5))
    plt.plot(df['year'], df[category], marker='o')
    plt.title(f'{category} over the Years')
    plt.xlabel('Year')
    plt.ylabel(category)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# Disasters affecting crops

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loaded variable 'df' from URI: c:\Users\jcbur\OneDrive - peruvianbusinesscouncil.org\csv files\#9_peru_agriculture\agriculture_disasters\natural_disasters_ha_may24.csv
df = pd.read_csv(r'C:\Users\jcbur\OneDrive - peruvianbusinesscouncil.com\csv files\#11_python_final_assignment\#3_bottomline_csv_ datasets\natural_disasters_ha_may24.csv')

# Filter rows based on column: 'region'
df = df[df['region'] == "Nacional"]

# Define categories for visualization
categories = ['crops_affected_by_disasters']

# Plotting each category
for category in categories:
    plt.figure(figsize=(10, 5))
    plt.plot(df['year'], df[category], marker='o')
    plt.title(f'{category} over the Years')
    plt.xlabel('Year')
    plt.ylabel(category)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
